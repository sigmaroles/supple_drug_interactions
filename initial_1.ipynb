```python
with open("./sentence_dict.json", "r") as fh:
    alltxt = fh.read()

print (f"{len(alltxt)} characters read from main data")
```

```python
with open("./paper_metadata.json", "r") as fh:
    alltxt2 = fh.read()

print (f"{len(alltxt2)} characters read from metadata")
```

## INFO
- Evidence sentences are located in `sentence_dict.json` --> this becomes `maindata`
- Paper metadata is provided in `paper_metadata.json` --> this becomes `metadata` 


```python
maindata = json.loads(alltxt)
metadata = json.loads(alltxt2)
```

```python
keys_inter = list(maindata.keys())
```

```python
keys_meta = list(metadata.keys())
```

```python
print (f"maindata keys look like {keys_inter[12]}")
print (f"metadata keys look like {keys_meta[12]}")

print (f"\nmetadata has {len(keys_meta)} items")
print (f"main data has {len(keys_inter)} items")
```

## linking maindata and metadata

```python
k1 = keys_inter[89]
md1 = maindata[k1] # md1 is a list with variable length -- at least one element
for indx, intr1 in enumerate(md1):
    print (indx+1)
    arg1 = intr1['arg1']
    arg2 = intr1['arg2']
    span_a2 = arg1['span']
    span_a1 = arg2['span']
    kw1 = intr1['sentence'][span_a1[0]: span_a1[1]]
    kw2 = intr1['sentence'][span_a2[0]: span_a2[1]]
    print (f"\"{kw1}\" <-> \"{kw2}\"")    
    print (f"\"{intr1['sentence']}\"")
    thismeta = metadata[intr1['paper_id']]
    print (f"({thismeta['year']}) {thismeta['title']}")
    print ("=============")
```
